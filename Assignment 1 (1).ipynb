```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filt = (salaries.employment_type == "FT")
max_salary = salaries[filt].groupby('job_title')['salary_in_usd'].max()
sorted_max_salary = max_salary.sort_values(ascending = False)
sorted_max_salary.plot(kind = 'bar',title='Highest FT salary',
               ylabel='Salary in USD', xlabel='Job title',figsize=(15,10))
```

Since the data deals with one categorical nomnal and numeric as the dependent variable, bar chart is the suitable graph
The data seems to have a great range of salary depending on a job. While the highest being the research scientist with the highest salary of approxiately 450000 USD while the the lowest was recorded by Power BI Developer with the salary of less than 10000 USD.

Justification: Since there are more than 6 categorie with each showing its discrete numerical data, the bar chart provides the cleanest and clear visibility

```python
company_location = salaries.groupby('company_location')['remote_ratio'].count()
company_location = company_location.sort_values(ascending = False)[:3]
company_location.plot(kind = "bar",title = 'Top three countries with the highest number of jobs', xlabel = 'Country', ylabel = "Frequency")
```

The top three countries with the highest number of records are US, GB and CA, with US being the most with 2500+ records

Jusification: alternative graph was pie chart, yet although that the number of data set shown here is 3, considering that there are many other countries recorded in the data, it is reasonable to use bar chart

```python
top_three_country = ['CA',"GB",'US']
new_salaries = salaries[(salaries.company_location == 'US') | (salaries.company_location == 'GB') | (salaries.company_location == 'CA')]
new_salaries = pd.DataFrame(new_salaries, columns=["company_location", "remote_ratio"])
remote_count1 = new_salaries[new_salaries.remote_ratio <20].groupby('company_location').count()
remote_count2 = new_salaries[(new_salaries.remote_ratio <80) & (new_salaries.remote_ratio >=20) ].groupby('company_location').count()
remote_count3 = new_salaries[new_salaries.remote_ratio >=80].groupby('company_location').count()
remote_count1['Partially_remote'] = remote_count2
remote_count1['Fully_remote'] = remote_count3
remote_count1.rename(columns={'remote_ratio':'No_remote'},inplace = True)
remote_count1 = remote_count1.reset_index()
remote_count1
remote_count1.plot(x="company_location", y=["No_remote", "Partially_remote","Fully_remote"], kind="bar", figsize=(9, 8))
```

From the graph shown, it is clear that majority of the jobs are either flly remote or no remote. In US and GB, no remote jobs are slightly more popular than fully remote job. while in CA, its an opposite case. Yet in all three countries, partially remote jobs are uncommon.

Justification of the graph: Multi graph is chosen in order to acheive the easiness in comparison between countries as well between remote ratio
